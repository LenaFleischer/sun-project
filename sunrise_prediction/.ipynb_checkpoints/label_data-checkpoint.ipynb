### Sunrise Date Dictionary
Read in the dates csv to make a dictionary of all dates with sunrise photos

```python
dates_path = os.path.join(os.getcwd(),"label_denver", "denver_sunrise.csv")
```

```python
df = pd.read_csv(dates_path)

dates = {}

# we only have one column here
for column in df: 
    columnSeriesObj = df[column]
    for item in columnSeriesObj.values:
        date = item[:10]
        # print(date)
        dates[date] = "sunrise"
```
